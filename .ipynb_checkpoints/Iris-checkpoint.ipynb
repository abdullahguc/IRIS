## Abstract: 

In this tutorial, we will be exploring the famous Iris dataset, looking for patterns in the dataset and classifying different spieces using the Support vector Machine algorithm.

## Data
The Iris data set covers three classes of flower spieces. Namely:
- **setosa**
- **versicolor**
- **virginica**

_**Note:**_ that when loading the dataset each class will be denoted by a number; 0, 1, 2 respectively.

Our data set contains 50 samples per class, which gives us a total of 150 samples. Each sample is characterized by 4 features all measured in centimeters. : 

1. **length of sepals.**
2. **width of sepals.**
3. **length of petals.**
4. **width of petals.**

where Petals are modified leaves that surround the reproductive parts of flowers and sepals are parts of the flower of angiosperms (flowering plants) according to wikipedia.


## Visualization
In this section we are going to graph and visualize trends in the Iris dataset 

### Loading the data
First thing we need to load the Iris data. Fortunately this can be easily done through the Scikit library allows to do this in just a few lines of code.

```python
# import iris dataset
from sklearn import datasets, svm
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, GridSearchCV


#load the dataset
(features, names) = datasets.load_iris(True)
# lets take a peak at the dataset
print(str(len(names)) + " samples were loaded successfully!")
```

### plotting the data

Now let's plot each feature and see if we can see trends to classify if each class and what type of curve we need to fit. Since it's very hard to visualize plots in 4D we will just visualize two features against each others to get a fealing of features relate to each other

```python
# purple for setosa, orange for versicolor and blue for virginica
colors = ['purple','orange','blue']

# sepals features plot
fig = plt.figure(figsize=(8,8))
plt.xlabel("length of sepals")
plt.ylabel("width of sepals")
plt.scatter(features[:,0], features[:,1], c=names, cmap=matplotlib.colors.ListedColormap(colors))

# petals features plot
fig = plt.figure(figsize=(8,8))
plt.xlabel("length of petals")
plt.ylabel("width of petals")
plt.scatter(features[:,2], features[:,3], c=names, cmap=matplotlib.colors.ListedColormap(colors))

```

# Support Vector Machine
Now let's train a support vector machine to be able to classify Iris flowers according to the four specified features.

## Training set and CV set 
first wee need to shuffle and split our data. We will do so using train_test_split() function defined in sklearn. and then use these sets to train and test our model. After doing so, we will perform a gridsearch to pick the best model with the most accuracy

```python
from sklearn import datasets, svm

(X_train, X_test, y_train, y_test) = train_test_split(features, names, test_size=0.3, random_state=42)

print("Training set size: " + str(len(X_train)) + " Test set size: " + str(len(X_test)) + "!")

# fit the model
#lin_clf = svm.LinearSVC()
#lin_clf.fit(X_train, y_train)

C_range = np.logspace(-2, 10, 15)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=42)
param_grid = dict(C=C_range)
linearsvm = GridSearchCV(svm.LinearSVC(), param_grid=param_grid, cv=cv)
linearsvm.fit(X_train, y_train)

print("The best parameters are %s with a score of %0.2f"
      % (linearsvm.best_params_, linearsvm.best_score_))
```

## Tryout some predictions !
Now that we have trained our model using the training model, let's use the test set and see how our model performs:

```python
print(linearsvm.predict(X_test))
```

## calculate the accuracy: 

```python
print('Accuracy: ', linearsvm.score(X_test, y_test))
```
